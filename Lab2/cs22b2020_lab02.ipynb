8. Handling Missing Values<br>
a) Check if there are any missing values in the DataFrame.<br>
b) If there are missing values, fill numeric columns with the mean of that
column and categorical columns with the mode (most frequent value) of that
column.

```python
missing_valeus = heart_failure.isnull().sum()
missing_valeus
```

```python
# Check for missing values again
if heart_failure.isnull().sum().any():
    # Separate numeric and categorical columns
    numeric_cols = heart_failure.select_dtypes(include=['float64', 'int64']).columns
    categorical_cols = heart_failure.select_dtypes(include=['object', 'category']).columns

    # Fill missing values in numeric columns with the mean
    heart_failure[numeric_cols] = heart_failure[numeric_cols].fillna(heart_failure[numeric_cols].mean())

    # Fill missing values in categorical columns with the mode (most frequent value)
    for col in categorical_cols:
        mode_value = heart_failure[col].mode()[0]  # Get the most frequent value (mode)
        heart_failure[col] = heart_failure[col].fillna(mode_value)

# Display the updated DataFrame with filled missing values
print("DataFrame after handling missing values:")
print(heart_failure.isnull().sum())  # Verify that there are no missing values now
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x values
x = np.linspace(0, 10, 100)

# Create a figure with 2 rows and 2 columns of subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle('Matplotlib Line Plot Customization', fontsize=16)

# First subplot (top-left)
axs[0, 0].plot(x, x, color='r', linewidth=0.5, label='y = x')  # Line 1
axs[0, 0].plot(x, x**2, color='g', linewidth=1, label='y = x^2')  # Line 2
axs[0, 0].plot(x, x**3, color='b', linewidth=2, label='y = x^3')  # Line 3
axs[0, 0].set_title('Basic Line Plots')
axs[0, 0].set_xlabel('x')
axs[0, 0].set_ylabel('y')
axs[0, 0].grid(True)
axs[0, 0].legend()

# Second subplot (top-right)
axs[0, 1].plot(x, np.sin(x), linestyle='-', label='Solid')      # Line 1
axs[0, 1].plot(x, np.sin(x), linestyle='--', label='Dashed')   # Line 2
axs[0, 1].plot(x, np.sin(x), linestyle='-.', label='Dash-dot')  # Line 3
axs[0, 1].plot(x, np.sin(x), linestyle=':', label='Dotted')     # Line 4
axs[0, 1].set_title('Line Styles')
axs[0, 1].set_xlabel('x')
axs[0, 1].set_ylabel('y')
axs[0, 1].grid(True)
axs[0, 1].legend()

# Third subplot (bottom-left)
markers = ['+', 'o', 's', '^']
sizes = [5, 8, 11, 14]
for marker, size in zip(markers, sizes):
    axs[1, 0].plot(x, np.cos(x), linestyle='-', marker=marker, markersize=size, label=f'Marker: {marker}')
axs[1, 0].set_title('Markers')
axs[1, 0].set_xlabel('x')
axs[1, 0].set_ylabel('y')
axs[1, 0].grid(True)
axs[1, 0].legend()

# Fourth subplot (bottom-right)
axs[1, 1].plot(x, np.tan(x), linestyle='--', color='m', marker='o', markerfacecolor='yellow', markeredgecolor='red', label='Custom')
axs[1, 1].set_title('Custom Line and Marker Styles')
axs[1, 1].set_xlabel('x')
axs[1, 1].set_ylabel('y')
axs[1, 1].grid(True)
axs[1, 1].legend()

# Adjust layout to prevent overlap
plt.tight_layout(rect=[0, 0, 1, 0.96])

# Save the figure as a PNG file
plt.savefig('matplotlib_assignment.png')

# Show the figure
plt.show()
```
